## 1.3 Regression

```python
from sklearn.model_selection import train_test_split

# Drop rows with missing values
beijing_reg_df = beijing_df.dropna()
chengdu_reg_df = chengdu_df.dropna()
guangzhou_reg_df = guangzhou_df.dropna()
shanghai_reg_df = shanghai_df.dropna()
shenyang_reg_df = shenyang_df.dropna()

# Treat the PM column as our predictive objective
bei_y = beijing_reg_df['PM']
che_y = chengdu_reg_df['PM']
gua_y = guangzhou_reg_df['PM']
sha_y = shanghai_reg_df['PM']
she_y = shenyang_reg_df['PM']

# All other columns will be used as features for training our model
bei_X = beijing_reg_df.drop(['PM'], axis=1)
che_X = chengdu_reg_df.drop(['PM'], axis=1)
gua_X = guangzhou_reg_df.drop(['PM'], axis=1)
sha_X = shanghai_reg_df.drop(['PM'], axis=1)
she_X = shenyang_reg_df.drop(['PM'], axis=1)

# Split 70% of the data for training and leave out 30% for testing
X_train_bei, X_test_bei, y_train_bei, y_test_bei = train_test_split(bei_X, bei_y, test_size=0.3, random_state=42)
X_train_che, X_test_che, y_train_che, y_test_che = train_test_split(che_X, che_y, test_size=0.3, random_state=42)
X_train_gua, X_test_gua, y_train_gua, y_test_gua = train_test_split(gua_X, gua_y, test_size=0.3, random_state=42)
X_train_sha, X_test_sha, y_train_sha, y_test_sha = train_test_split(sha_X, sha_y, test_size=0.3, random_state=42)
X_train_she, X_test_she, y_train_she, y_test_she = train_test_split(she_X, she_y, test_size=0.3, random_state=42)
```

### 1.3.1 Train and evaluate (using r-squared) a linear regression model to predict the PM value. Repeat this for each city and be sure to evaluate your model using the X test set.

```python
from sklearn.linear_model import LinearRegression

# Train Beijing
lr = LinearRegression()
lr.fit(X_train_bei, y_train_bei)
#Predict
predict_bei = lr.predict(X_test_bei)
predict_bei
```

```python
from sklearn.metrics import r2_score

# Evaluate prediction
true_bei = y_test_bei
pred_bei = predict_bei
linear_bei = r2_score(true_bei, pred_bei)
linear_bei
```

```python
# Train Chengdu
lr.fit(X_train_che, y_train_che)
# Predict
predict_che = lr.predict(X_test_che)
# Evaluate prediction
true_che = y_test_che
pred_che = predict_che
linear_che = r2_score(true_che, pred_che)
linear_che
```

```python
# Train Guangzhou
lr.fit(X_train_gua, y_train_gua)
# Predict
predict_gua = lr.predict(X_test_gua)
# Evaluate prediction
true_gua = y_test_gua
pred_gua = predict_gua
linear_gua = r2_score(true_gua, pred_gua)
linear_gua
```

```python
# Train Shanghai
lr.fit(X_train_sha, y_train_sha)
# Predict
predict_sha = lr.predict(X_test_sha)
# Evaluate prediction
true_sha = y_test_sha
pred_sha = predict_sha
linear_sha = r2_score(true_sha, pred_sha)
linear_sha
```

```python
# Train Shenyang
lr.fit(X_train_she, y_train_she)
# Predict
predict_she = lr.predict(X_test_she)
# Evaluate prediction
true_she = y_test_she
pred_she = predict_she
linear_she = r2_score(true_she, pred_she)
linear_she
```
