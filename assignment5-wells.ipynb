```python
import pandas as pd

# Load data for all 5 cities
path = "../data/"
beijing_df = pd.read_csv(path + "Beijing.txt")
chengdu_df = pd.read_csv(path + "Chengdu.txt")
guangzhou_df = pd.read_csv(path + "Guangzhou.txt")
shanghai_df = pd.read_csv(path + "Shanghai.txt")
shenyang_df = pd.read_csv(path + "Shenyang.txt")
```

- What are the hottest summer and coldest winters observed? Explain how you defined hottest and coldest?

**I will define hottest summer as the summer (by year) having the most instances of average high temperature and humidity in the top 10. I chose this because a max temperature could be an outlier and not as representative, and I chose both temperature and humidity because humidity intensifies the feeling of being hot. For coldest winter, I will use only temperature as there is debate regarding how humidity affects the feeling of being cold. The other measure is the same: the winter (by year) having the most instances of average low temperatures in the top 10.**

- Which feature appears to contain the largest amount of missing data overall?

```python
# Print the total number of null instances for each Beijing feature
for column in beijing_df:
    i = beijing_df[column]
    null_sum = (i.isnull() == True).sum()
    print("Number of null " + column + " instances: ")
    print(null_sum)
```

```python
# Print the total number of null instances for each Chengdu feature
for column in beijing_df:
    i = chengdu_df[column]
    null_sum = (i.isnull() == True).sum()
    print("Number of null " + column + " instances: ")
    print(null_sum)
```

```python
# Print the total number of null instances for each Guangzhou feature
for column in beijing_df:
    i = guangzhou_df[column]
    null_sum = (i.isnull() == True).sum()
    print("Number of null " + column + " instances: ")
    print(null_sum)
```

```python
# Print the total number of null instances for each Shanghai feature
for column in beijing_df:
    i = shanghai_df[column]
    null_sum = (i.isnull() == True).sum()
    print("Number of null " + column + " instances: ")
    print(null_sum)
```

```python
# Print the total number of null instances for each Shenyang feature
for column in beijing_df:
    i = shenyang_df[column]
    null_sum = (i.isnull() == True).sum()
    print("Number of null " + column + " instances: ")
    print(null_sum)
```

**For each individual city as well as all cities overall, the PM feature appears to contain the largest amount of missing data. Even a city such as Guangzhou, which has very few null values (8 total not including the PM feature), has 20,232 null PM values.**
